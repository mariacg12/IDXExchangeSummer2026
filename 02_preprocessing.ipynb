Make new data frames using a copy of housing

```python
housing_drop = housing.copy()
housing_median = housing.copy()
housing_feature = housing.copy()
```

1. Dropping all missing and duplicate values

```python
housing_drop = housing_drop.drop_duplicates(subset="ListingKey")
housing_drop = housing_drop.dropna()
```

2. Using median and mode

```python
housing_median = housing_median.drop_duplicates(subset="ListingKey")

num_cols = housing_median.select_dtypes(include=np.number).columns
catg_cols = housing_median.select_dtypes(include="object").columns

housing_median[num_cols] = housing_median[num_cols].fillna(housing_median[num_cols].median())

for col in catg_cols:
    housing_median[col] = housing_median[col].fillna(housing_median[col].mode()[0])
```

```python
missing = housing_median.isnull().sum()
print("missing:\n",missing)

print("\nduplicates:")
dupes = housing_median["ListingKey"].duplicated().sum()
print(dupes)

print(f"Rows kept: {len(housing_median)/len(housing)*100:.2f}%")
```

Columns being dropped

```python
drop_cols = [
    "OriginalListPrice",
    "ListPrice",
    "ListingKeyNumeric",
    "ListingId",
    "ListAgentFirstName",
    "ListAgentLastName",
    "ListAgentFullName",
    "ListAgentEmail",
    "BuyerAgentFirstName",
    "BuyerAgentLastName",
    "BuyerAgentMlsId",
    "CoBuyerAgentFirstName",
    "CoListAgentFirstName",
    "CoListAgentLastName",
    "ListOfficeName",
    "BuyerOfficeName",
    "CoListOfficeName",
    "UnparsedAddress"]

#dropped more after seeing these columns had a great amount of missing values
drop_cols.extend([
    "TaxAnnualAmount",
    "AboveGradeFinishedArea",
    "FireplacesTotal",
    "CoveredSpaces",
    "TaxYear",
    "StreetNumberNumeric",
    "PurchaseContractDate"])
```

Numeric Columns that will be inputed with median

```python
median_cols = [
    "LivingArea",
    "DaysOnMarket",
    "FireplacesTotal",
    "AboveGradeFinishedArea",
    "TaxAnnualAmount",
    "ParkingTotal",
    "LotSizeAcres",
    "YearBuilt",
    "BathroomsTotalInteger",
    "BuildingAreaTotal",
    "BedroomsTotal",
    "BelowGradeFinishedArea",
    "CoveredSpaces",
    "Stories",
    "LotSizeArea",
    "MainLevelBedrooms",
    "GarageSpaces",
    "AssociationFee",
    "Latitude",
    "Longitude",
    "LotSizeSquareFeet"]
```

Categorical columns that will be inputed with mode

```python
mode_cols = [
    "BuyerAgentAOR",
    "ListAgentAOR",
    "AssociationFeeFrequency",
    "MLSAreaMajor",
    "CountyOrParish",
    "MlsStatus",
    "ElementarySchool",
    "ElementarySchoolDistrict",
    "BuilderName",
    "SubdivisionName",
    "BuyerOfficeAOR",
    "BusinessType",
    "StateOrProvince",
    "MiddleOrJuniorSchool",
    "MiddleOrJuniorSchoolDistrict",
    "HighSchool",
    "HighSchoolDistrict",
    "City",
    "PostalCode",
    "Levels",
    "LotSizeDimensions"]
```

Boolean

```python
unknown_cols = [
    "ViewYN",
    "WaterfrontYN",
    "BasementYN",
    "PoolPrivateYN",
    "AttachedGarageYN",
    "FireplaceYN",
    "NewConstructionYN"]
```

Dropping columns and duplicates, and remove missing and null values from close price

```python
housing_feature = housing_feature.drop(columns=drop_cols)
housing_feature = housing_feature.drop_duplicates(subset="ListingKey")
housing_feature = housing_feature.dropna(subset=["ClosePrice"])
```

Imputation using median and model and adding unknown to boolean columns

```python
for col in median_cols:
    if col in housing_feature.columns:
        housing_feature[col] = housing_feature[col].fillna(housing_feature[col].median())

for col in mode_cols:
    if col in housing_feature.columns:
        # Check if mode returns an empty Series
        if not housing_feature[col].mode().empty:
            housing_feature[col] = housing_feature[col].fillna(housing_feature[col].mode()[0])
        else:
            # If mode fill with Unknown
            housing_feature[col] = housing_feature[col].fillna("Unknown")

for col in unknown_cols:
    if col in housing_feature.columns:
        housing_feature[col] = housing_feature[col].fillna("Unknown")

housing_feature["Flooring"] = housing_feature["Flooring"].fillna("Unknown")
```

```python
missing = housing_feature.isnull().sum()
print("missing:\n",missing)

print("\nduplicates:")
dupes = housing_feature["ListingKey"].duplicated().sum()
print(dupes)

print(f"Rows kept: {len(housing_feature)/len(housing)*100:.2f}%")

housing_feature.isnull().sum().sort_values(ascending=False).head(20)
```

```python
comparison = pd.DataFrame({
    "Strategy": [
        "Original",
        "Drop all missing",
        "Median/Mode",
        "Feature-specific"],
    "Rows": [
        len(housing),
        len(housing_drop),
        len(housing_median),
        len(housing_feature)],
    "Columns": [
        housing.shape[1],
        housing_drop.shape[1],
        housing_median.shape[1],
        housing_feature.shape[1]],
    "Missing Values": [
        housing.isnull().sum().sum(),
        housing_drop.isnull().sum().sum(),
        housing_median.isnull().sum().sum(),
        housing_feature.isnull().sum().sum()]})

comparison["Rows Kept %"] = (comparison["Rows"] / len(housing) * 100).round(2)
comparison["Columns Kept %"] = (comparison["Columns"] / housing.shape[1] * 100).round(2)
comparison["Missing Values %"] = (comparison["Missing Values"] / housing.isnull().sum().sum() * 100).round(2)

print(comparison)

```

Use the most recent month as the test set and the res as the training set

```python
#time base train/test split
latest_month = housing_encoded["DataMonth"].max()

train = housing_encoded[housing_encoded["DataMonth"] < latest_month]
test  = housing_encoded[housing_encoded["DataMonth"] == latest_month]

X_train = train.drop(columns=["ClosePrice"])
y_train = train["ClosePrice"]

X_test = test.drop(columns=["ClosePrice"])
y_test = test["ClosePrice"]

# Remove DataMonth after splitting since it is no longer needed as a predictor
X_train = X_train.drop(columns=["DataMonth"])
X_test = X_test.drop(columns=["DataMonth"])
```
