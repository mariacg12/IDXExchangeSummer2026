Load data set

```python
df_202505 = pd.read_csv("CRMLSSold202505.csv")
df_202506 = pd.read_csv("CRMLSSold202506.csv")
df_202507 = pd.read_csv("CRMLSSold202507.csv")
df_202508 = pd.read_csv("CRMLSSold202508.csv")
df_202509 = pd.read_csv("CRMLSSold202509.csv")
df_202510 = pd.read_csv("CRMLSSold202510.csv")
df_202511 = pd.read_csv("CRMLSSold202511.csv")
df_202512 = pd.read_csv("CRMLSSold202512.csv")
df_202601 = pd.read_csv("CRMLSSold202601.csv")
df_202602 = pd.read_csv("CRMLSSold202602.csv")
df_202603 = pd.read_csv("CRMLSSold202603.csv")
df_202604 = pd.read_csv("CRMLSSold202604.csv")
df_202605 = pd.read_csv("CRMLSSold202605.csv")
```

distribution of close price

```python
sns.set_theme(
    style="whitegrid",
    palette="Set2",
    font_scale=1.1)

plt.figure(figsize=(10,6))

sns.histplot(
    housing["ClosePrice"],
    bins=50,
    kde=True)

plt.title("Distribution of Close Price")
plt.xlabel("Close Price")
plt.ylabel("Number of Properties")

plt.show()
```

Using log prices instead because the close price distribution is skewed probably due to outliars. Using log we can have a better unddrtanding of the distribution

---



```python
sns.set_theme(
    style="whitegrid",
    palette="Set2",
    font_scale=1.1)

plt.figure(figsize=(10,6))

housing["LogClosePrice"] = np.log(housing["ClosePrice"])

sns.histplot(
    housing["LogClosePrice"],
    bins=50,
    kde=True)

plt.title("Distribution of Log Close Price")
plt.xlabel("Close Price")
plt.ylabel("Number of Properties")

plt.show()
```
